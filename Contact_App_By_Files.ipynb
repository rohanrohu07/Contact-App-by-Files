```python
'''To run this code, make sure to create a file called contacts.txt in your operating system and store
its file path in the contacts variable before running the program.'''

import re
def name_check(name):
  r=re.match(r'[A-Z,a-z]',name)
  if r == name:
    return True
  else:
    return False

def number_check(number):
  r=re.match(r'\d{10}',number)
  if r == number:
    return True
  else:
    return False

def email_check(email):
  r=re.match(r'\w+\.?\w+@\w+\.[a-z]{3}',email)
  if r == email:
    return True
  else:
    return False
def add_contacts(contacts):
  name = input('Enter The Name Of The New Contact: ')
  name_check(name)
  number = input('Enter The Phone Number Of The New Contact: ')
  number_check(number)
  email = input('Enter The Email Of The New Contact: ')
  email_check(email)
  if name == name_check(name) and number == number_check(name) and email == email_check(email):
    with open(contacts,'a') as f: 
      new_contact=name + ','+ number + ',' + email 
      if f.tell() == 0:
        f.write(new_contact)
      else:
        f.write('\n'+new_contact)


def open_contact(contacts,name):
  with open(contacts) as f:
    f1=f.readlines()
    x=''.join(f1)
    for contact in f1:
      id=f1.index(contact)
      o=contact
      contact=contact.split(',')
      if contact[0].lower() == name.lower():
        choice = input("Press 'u' to update or 'd' to delete the contact: ")
        if choice.lower() == 'u':
          new_name = input("Enter new name (or press enter to keep current name): ")
          name_check(new_name)
          if new_name == name_check(new_name):
            if new_name:
              contact[0]='\n'+new_name
          new_number = input("Enter new number (or press enter to keep current number): ")
          number_check(new_number)
          if new_number == number_check(new_number):
            if new_number:
              contact[1]=new_number
          new_email = input("Enter new email (or press enter to keep current email): ")
          email_check(new_email)
          if new_email == email_check(new_email):
            if new_email:
              contact[2]=new_email
          f1[id]=','.join(contact)
          print("Contact updated successfully.")
          print(f1)
          x=''.join(f1)
          break
        elif choice.lower() == 'd':
          f1.remove(o)
          print("Contact deleted successfully.")
          x=''.join(f1)
          break
    else:
      print("No match found!")
  with open(contacts,'w') as f:
    f.write(x)

def search_contact(contacts,search_term):
  with open(contacts) as f:
    f1=f.readlines()
    results=[]
    for contact in f1:
      contact=contact.split(',')
      if search_term.lower() in contact[0].lower():
        results.append(contact)
    if len(results) > 0:
      print("Search results:")
      for result in results:
        print(result[0])
    else:
      print("No matching contacts found.")


while True:
  contacts='contacts.txt'
  print('''*******Welcome to The Contact App*******
          1.Add Contact
          2.Open Contact
          3.Search Contact
          4.Quit''')
  choice = input('\nEnter Your Choice : ')
  if choice == '1':
    add_contacts(contacts)
  elif choice == '2':
    name = input("Enter name of the contact you want to open. ")        
    open_contact(contacts, name)
  elif choice == '3':
    search_term = input("Enter search term: ")
    search_contact(contacts, search_term)
  elif choice == '4':
    break
  else:
    print("Incorrect choice.")
```
